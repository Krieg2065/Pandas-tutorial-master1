```python
%matplotlib inline
import numpy as np
import pandas as pd
import squarify    # pip install squarify (algorithm for treemap)
import matplotlib.pyplot as plt

pd.options.display.max_rows = 8
```

```python
df = pd.read_csv("data/vaccini2.csv")
df
```

```python
area = df['area'].head(21)
dosi = df['dosi_somministrate'].head(21)
fig, ax = plt.subplots(figsize =(16, 9))
ax.barh(area, dosi)
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
ax.grid(b = True, color ='grey',linestyle ='-.', linewidth = 0.5,alpha = 0.2)
ax.set_title('Vaccini',loc ='left', )
plt.show()
```

```python
X = list(df.iloc[:, 0])
Y = list(df.iloc[:, 1])
fig, ax = plt.subplots(figsize =(16, 9))
sorted(X)
X.reverse()
plt.barh(X, Y)
plt.title("Vaccini")
plt.xlabel("Area")
plt.ylabel("Dosi somministrate")
plt.show()
```

```python
area = df['area'].head(21)
dosi = df['percentuale_somministrazione'].head(21)
fig, ax = plt.subplots(figsize =(16, 9))
ax.bar(area, dosi)
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
ax.grid(b = True, color ='grey',linestyle ='-.', linewidth = 0.5,alpha = 0.2)
ax.set_title('Vaccini',loc ='left', )
ax.set_ylim(70, 100)
plt.show()
```

```python
data = [df['dosi_somministrate'],
df['dosi_consegnate']]
X = np.arange(21)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
```

```python
labels = df['area']
sizes = df['dosi_somministrate']


fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```
